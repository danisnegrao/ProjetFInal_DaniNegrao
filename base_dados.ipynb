***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Extraíndo algumas colunas***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python
columns_drop = ['Carimbo de data/hora','Como Gostaria de ser chamada?','Já teve filhos?']
df_dados2 = df_dados.drop(columns_drop, axis=1)
df_dados2
```

```python

df_dados2.rename(columns={'Quantos filhos você tem? Caso não tenha pode colocar 0.':'Quantidade_Filhos'}, inplace=True)
df_dados2
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Contagem da Idade das 223 mulheres entrevistadas***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python
contagem_idades = df_dados2['IDADE:'].value_counts().sort_index()
contagem_idades
```

```python

fig, ax = plt.subplots()
barras = ax.barh(contagem_idades.index, contagem_idades.values, color='purple')
ax.barh(contagem_idades.index, contagem_idades.values, color='purple')

for barra in barras:
    xval = barra.get_width()
    ax.text(xval, barra.get_y() + barra.get_height()/2, int(xval), ha='left', va='center')


plt.title('Distribuição de Idades')
plt.xlabel('Número de Mulheres')
plt.ylabel('IDADE')
plt.grid(axis='y', linestyle='--', alpha=0.1)

plt.show()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Média de filhos por Idade***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python
media_filhos_por_idade = df_dados2.groupby('IDADE:')['Quantidade_Filhos'].mean()
print(media_filhos_por_idade)
```

```python
plt.barh(media_filhos_por_idade.index, media_filhos_por_idade.values, color='purple')
plt.title('Média de Filhos por Idade')
plt.xlabel('Idade')
plt.ylabel('Média de Filhos')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Métodos Contraceptivos Utilizados***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python
contagem_metodo_contraceptivo = df_dados2['Qual o método contraceptivo você utiliza atualmente?'].value_counts().sort_index()
contagem_metodo_contraceptivo
```

```python
fig, ax = plt.subplots()
barras = ax.barh(contagem_metodo_contraceptivo.index, contagem_metodo_contraceptivo.values, color='purple')
ax.barh(contagem_metodo_contraceptivo.index, contagem_metodo_contraceptivo.values, color='purple')

for barra in barras:
    xval = barra.get_width()
    ax.text(xval, barra.get_y() + barra.get_height()/2, int(xval), ha='left', va='center')


plt.title('Métodos Contraceptivos Utilizado')
plt.xlabel('Número de Mulheres')
plt.ylabel('Método')
plt.grid(axis='y', linestyle='--', alpha=0.1)

plt.show()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Frequência de Orgasmos***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python
freq_orgasmos = df_dados2['Com que frequência você atinge o orgasmo durante suas experiências sexuais?'].value_counts().sort_index()
freq_orgasmos
```

```python
fig, ax = plt.subplots()
barras = ax.barh(freq_orgasmos.index, freq_orgasmos.values, color='purple')
ax.barh(freq_orgasmos.index, freq_orgasmos.values, color='purple')

for barra in barras:
    xval = barra.get_width()
    ax.text(xval, barra.get_y() + barra.get_height()/2, int(xval), ha='left', va='center')


plt.title('Frequência de Orgasmos')
plt.xlabel('Número de Mulheres')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.1)

plt.show()
```

***<div style="background-color:white; padding:10px;text-align:center"><span style="color:purple; font-size:2em">Relação entre Frequência de Orgasmos e a Intimidade com seu Própio Corpo***<img src="https://img1.picmix.com/output/stamp/normal/6/7/4/6/1606476_e9643.gif" alt="" width="50"/></span></div>


```python


mapeamento_sensacoes = {'Muito confortável': 5, 'Confortável': 4, 'Neutro': 3, 'Desconfortável': 2, 'Muito desconfortável': 1}
mapeamento_frequencia = {'Raramente': 2, 'Ocasionalmente': 3, 'Frequentemente': 4, 'Sempre': 5, 'Nunca':1 }

df_dados2['Sensacoes_Ao_Tocar_Ordenado'] = df_dados2['Como você se sente em relação a tocar seu corpo?'].map(mapeamento_sensacoes)
df_dados2['Frequencia_Orgasmos_Ordenado'] = df_dados2['Com que frequência você atinge o orgasmo durante suas experiências sexuais?'].map(mapeamento_frequencia)

plt.scatter(x=df_dados2['Sensacoes_Ao_Tocar_Ordenado'], y=df_dados2['Frequencia_Orgasmos_Ordenado'], c=df_dados2['Frequencia_Orgasmos_Ordenado'], cmap='viridis', edgecolor='k', s=100)


plt.title('Relação entre Sensações ao se Tocar e Frequência de Orgasmos')
plt.xlabel('Sensações ao se Tocar')
plt.ylabel('Frequência de Orgasmos')

cbar = plt.colorbar()
cbar.set_label('Frequência de Orgasmos')


plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Sensacoes_Ao_Tocar_Ordenado', hue='Com que frequência você atinge o orgasmo durante suas experiências sexuais?', data=df_dados2, palette='viridis')


plt.title('Relação entre Sensações ao se Tocar e Frequência de Orgasmos')
plt.xlabel('Sensações ao se Tocar')
plt.ylabel('Quantidade de Pessoas')


legend_labels_sensacoes = {1: 'Muito confortável', 2: 'Confortável', 3: 'Neutro', 4: 'Desconfortável', 5: 'Muito desconfortável'}
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=[f'{k}: {v}' for k, v in legend_labels_sensacoes.items()], title='Sensações ao se Tocar')


ax.legend(title='Frequência de Orgasmos', loc='upper left')


plt.show()
```

```python


desvio_padrao_orgasmos = np.std(df_dados2['Frequencia_Orgasmos_Ordenado'])
print(f'Desvio Padrão da Frequência de Orgasmos: {desvio_padrao_orgasmos:.2f}')
```
